```python
# Convert BigQuery data date fields from datetime.date objects to strings
bigquery_data2 = []
for row in query_job.result():
    # Convert date fields (if they are datetime.date) to a formatted string
    converted_row = []
    for col in row:
        if isinstance(col, datetime.date):
            converted_row.append(col.strftime('%Y-%m-%d'))
        else:
            converted_row.append(col)
    bigquery_data2.append(tuple(converted_row))

# Print converted BigQuery data
for row in bigquery_data2:
    print(row)
```

```python
# Function to compare SQLite and BigQuery schemas
def compare_schemas(sqlite_schema, bigquery_schema):
    sqlite_columns = {col[1]: col[2] for col in sqlite_schema}  # {column_name: data_type}
    bigquery_columns = {field.name: field.field_type for field in bigquery_schema}  # {column_name: data_type}

    # Compare columns
    for column in sqlite_columns:
        if column in bigquery_columns:
            if sqlite_columns[column].upper() != bigquery_columns[column].upper():
                print(f"Data type mismatch in column '{column}': SQLite({sqlite_columns[column]}) vs BigQuery({bigquery_columns[column]})")
        else:
            print(f"Column '{column}' exists in SQLite but not in BigQuery")

    # Check if BigQuery has extra columns
    for column in bigquery_columns:
        if column not in sqlite_columns:
            print(f"Column '{column}' exists in BigQuery but not in SQLite")

# Call the comparison function
compare_schemas(sqlite_schema, bigquery_schema)
```

```python
# Call the comparison function for sql table2
compare_schemas(sqlite_schema_2, bigquery_schema)
```

```python
def largest(a,b):
  if list(a.values())[0] > list(b.values())[0]:
    return a
  else:
    return b
```

```python
def smallest(a,b):
  if list(a.values())[0] < list(b.values())[0]:
    return a
  else:
    return b
```

```python
def diff(a,b):
  return list(largest(a,b).values())[0] - list(smallest(a,b).values())[0]
```
